## Discrete Probability Distribution Types (Designed by Probability Mass Functions)

### Uniform Discrete Probability Distribution

```python
# Throw a Dice Once
# Discrete Uniform Probability Distribution of A Dice (PMF)
a=1
b=6

x = numpy.arange(a, b+1)

print(x)
discrete_uniform_distribution = randint(a, b+1)
discrete_uniform_pmf = discrete_uniform_distribution.pmf(x)

print(discrete_uniform_pmf)
pyplot.plot(x, discrete_uniform_pmf, 'bo', ms=8)
pyplot.vlines(x, 0, discrete_uniform_pmf, colors='b', lw=5, alpha=0.5)
pyplot.xlabel('X axis is Sides of the Ludo Dice')
pyplot.ylabel('Probability')
pyplot.show()
```

#### Cumulatime Probability Distribution of the Above

```python
# Cumulatime Probability Distribution of the Above
discrete_uniform_cdf = discrete_uniform_distribution.cdf(x)
pyplot.plot(x, discrete_uniform_cdf, 'bo', ms=8)
pyplot.xlabel('X axis is Sides of the Ludo Dice')
pyplot.ylabel('Cumulative Probability')
pyplot.show()
```

### Negative Binomial Distribution
![title](DataFile/Neg.png)

```python
from scipy.stats import nbinom
#
# X = Discrete negative binomial random variable representing number of sales call required to get r=3 leads
# P = Probability of successful sales call
#
X = numpy.arange(3, 30)
r = 3
P = 0.1
#
# Calculate geometric probability distribution
#
nbinom_pd = nbinom.pmf(X, r, P)
#
# Plot the probability distribution
#
fig, ax = pyplot.subplots(1, 1, figsize=(8, 6))
ax.plot(X, nbinom_pd, 'bo', ms=8, label='nbinom pmf')
ax.plot(X, nbinom_pd, 'bo', ms=8, label='nbinom pmf')
pyplot.ylabel("Probability", fontsize="18")
pyplot.xlabel("X - No. of Sales Call", fontsize="18")
pyplot.title("Negative Binomial Distribution - No. of Sales Call Vs Probability", fontsize="18")
ax.vlines(X, 0, nbinom_pd, colors='b', lw=5, alpha=0.5)
```

## Continuous Probability Distribution Types (Designed by Probability Density Functions, Calculus for Anti Derivative Area)
* Probability Density Functions is Almost Like Histogram but the Histogram Bar Group Interval is Extremely thin & almost near to Zero & that's why we use Calculus for Anti Derivative Area as Probability.

### Continuous Uniform Probability Distribution

```python
from scipy.stats import uniform
a=0
b=20
size=4000

x = numpy.linspace(a, b, size)

print(x[:3])
continuous_uniform_distribution = uniform(loc=a, scale=b)
continuous_uniform_pdf = continuous_uniform_distribution.pdf(x)

print(continuous_uniform_pdf)
pyplot.plot(x, continuous_uniform_pdf)
pyplot.xlabel('X')
pyplot.ylabel('Probability ')
pyplot.show()
```

### Continuous Normal Probability Distribution Alongwith It's Cumulative Probability Distribution
* For Continuous Normal Probability Distribution, Area Means Cumulative Density.
![title](DataFile/Normal.png)

```python
from IPython.display import Image
Image(filename="DataFile/Norm.jpg", width=900, height=900)
```

```python
# pdf and cdf for a normal distribution
from scipy.stats import norm
from matplotlib import pyplot
# define distribution parameters
mu = 50
sigma = 5
# create distribution
dist = norm(mu, sigma)
# plot pdf
values = [value for value in range(30, 70)]
probabilities = [dist.pdf(value) for value in values]
pyplot.plot(values, probabilities)
pyplot.ylabel('Y axis is Likelyhood')
pyplot.show()
# plot cdf
cprobs = [dist.cdf(value) for value in values]
pyplot.plot(values, cprobs)
pyplot.ylabel('Y axis is Left Side Probability Density')
pyplot.show()
```

### Continuous Exponential Probability Distribution Alongwith Cumulative Probability

```python
# pdf and cdf for an exponential distribution
from scipy.stats import expon
# define distribution parameter
beta = 50
# create distribution
dist = expon(beta)
# plot pdf
values = [value for value in range(50, 70)]
probabilities = [dist.pdf(value) for value in values]
pyplot.plot(values, probabilities)
pyplot.show()
# plot cdf
cprobs = [dist.cdf(value) for value in values]
pyplot.plot(values, cprobs)
pyplot.show()
```
