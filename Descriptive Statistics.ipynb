# Types of Probability Distribution
1. Discrete Random Variable Probability Distribution
* Discrete Random Variables can take on a countable number of Possible Values
* We use Summation of Probability Mass Function
* Probability for an Exact Value of a Random Variable is Possible.
* Equation is Called PMF ie Probability Mass Function & P(x) >= 0, Î£P(x) = 1

2. Continuous Random Variable Probability Distribution
* We use intigration of Likelyhood Probability Density Function
* A random variable is continuous if it can be described by a Likelyhood Probability Density Function
* Probability for an Exact Value of Continuous Random Variable is not Possible
* Equation is Called Likelyhood PDF ie Probability Density Function $ \int_{-\infty}^{\infty} f(x) \,dx = 1 $
* f(x) likelyhood output value at any point of Continuous Random Variable x value, always will be $ f(x) >= 0 $

## Discrete Random Variable Probability Types

### Uniform & Cumulative Discrete Probability Distribution
* Throw a Dice Once
* Uniform Probability Distribution of A Dice (PMF)
* Cumulative Probability Distribution of A Dice (CDF)
* Die Roll is equivalent to Successively (One by one) With Replacement Method

```python
import numpy as np
from scipy.stats import randint
import matplotlib.pyplot as plt

a, b = 1, 6
x = np.arange(a, b + 1)
dist = randint(a, b + 1)

pmf = dist.pmf(x)
cdf = dist.cdf(x)

# PMF plot
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, pmf, 'bo', ms=10, label='PMF')
plt.vlines(x, 0, pmf, colors='b', lw=4, alpha=0.7)
plt.title('PMF of Fair Six-Sided Die')
plt.xlabel('X axis is Sides of the Ludo Dice')
plt.ylabel('Y axis is Uniform Probability')
plt.xticks(x)
plt.ylim(0, 0.2)
plt.grid(True, alpha=0.3)
plt.legend()

# CDF plot (step style)
plt.subplot(1, 2, 2)
# Plot steps: horizontal line up to the point, then jump
plt.hlines(cdf, xmin=np.roll(x, 1), xmax=x, color='blue', alpha=0.7)
plt.plot(x, cdf, 'bo', ms=8)
plt.vlines(x, ymin=np.roll(cdf, 1), ymax=cdf, color='blue', alpha=0.7)

# Add the starting point at 0
plt.hlines(0, 0, 1, color='blue', alpha=0.7)
plt.title('CDF of Fair Six-Sided Die')
plt.xlabel('X axis is Sides of the Ludo Dice')
plt.ylabel('Y axis is Cumulative Probability')
plt.xticks(x)
plt.ylim(0, 1.05)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import math

# Mean of the Discrete Uniform Probability Distribution
# E(x) = Î£ x p(x)

Mean = sum((x * 1/6 ) for x in x)
print(f"Mean is {Mean}")      # means if we throw a Ludo Dice 1 time, on average 3.5 will come

# Var(x) = E(x^2) - [E(x)]^2
Var = sum(((x ** 2) * 1/6 ) for x in x) - (Mean) ** 2
print(f"Variance is {Var:.2f}")

# Standard Deviation
Std_Dev = math.sqrt(Var)
print(f"Std_Dev is {Std_Dev:.2f}")
```

### Binomial ( 2 outcomes success & failure ) Distribution & it's Cumulative Probability Distribution
* Coin Toss is equivalent to Successively (One by one) With Replacement Method
* Tossing a Dice for Appearing 6 or any other particular side as success is equivalent to Successively (One by one) With Replacement Method

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# --- Parameters ---
n = 10  # Total number of trials (e.g., coin flips)
p = 0.5 # Probability of success (e.g., getting heads)

# --- Calculate PMF for all possible outcomes (k = 0 to n) ---
# k: The number of successful outcomes you're interested in
k_values = list(range(n + 1))
probabilities = [binom.pmf(k, n, p) for k in k_values]

# --- Display Results ---
print(f"Binomial Probabilities for n={n}, p={p}:")
for k, prob in zip(k_values, probabilities):
    print(f"P(k={k}): {prob:.4f}")

# --- Optional: Plot the distribution (requires matplotlib) ---
plt.bar(k_values, probabilities)
plt.xlabel("X - axis is Number of Times Head (Successes (k)) Appears")
plt.ylabel("Y - axis is Binomial Probability")
plt.title(f"Binomial Distribution PMF (n={n}, p={p})")
plt.show()
```

```python
# Cumulatime Probability Distribution of the Above
Cumulative_Prob_dist = [binom.cdf(k, n, p) for k in k_values]

# --- Display Results ---
print(f"Binomial Cumulative Probabilities for n={n}, p={p}:")
for k, Cum_prob in zip(k_values, Cumulative_Prob_dist):
    print(f"P(k={k}): {Cum_prob:.4f}")

# plotting the graph
plt.ylabel('Y - axis is Binomial Cumulative Probability')
plt.xlabel('X - axis is Upto no of times Head (Successes (k)) Appears')
plt.title(f"Binomial Distribution CDF (n={n}, p={p})")
plt.bar(k_values, Cumulative_Prob_dist)
plt.show()
```

```python
# Mean of the Discrete Binomial Probability Distribution
# Mean(Î¼) = n * p

Mean = n * p
print(f"Mean is {Mean}")      # means if we Toss a Coin 10 times, on average 5 times Head will come

# Variance of the Discrete Binomial Probability Distribution
# The variance of a discrete probability distribution measures the spread of outcomes, calculated as the 
# Variance = n * p * q
Variance = n * p * (1 - p)
print(f"Variance is {Variance:.2f}")

# Standard Deviation
Std_Dev = math.sqrt(Variance)
print(f"Std_Dev is {Std_Dev:.2f}")
```

### Poisson Discrete Probability Distribution
* Poison Distribution is a kind of Binomial Distribution (Successively (One by one) With Replacement)
* Where No of Trials are greater than 20
* Distribution mean Î¼ = np
* Variance (ÏÂ²) = np
* e = 2.71828
* n = number of trials
* p = prob of success is very small
* q = prob of failure is very big
* X = required no of success

<img src="DataFile/poisson.png"/>
<img src="DataFile/Poisson Binom Relation.png"/>

```python
import math
import matplotlib.pyplot as plt

def poisson_pmf(k, lam):
    """
    Calculate Poisson probability mass function (PMF).
    
    P(X = k) = (Î»^k * e^(-Î»)) / k!
    
    Parameters:
    k: number of events (non-negative integer)
    lam: average rate (lambda parameter, positive)
    
    Returns:
    Probability of exactly k events
    """
    if k < 0 or not isinstance(k, int):
        return 0
    if lam <= 0:
        raise ValueError("Lambda must be positive")
    
    return (lam ** k) * math.exp(-lam) / math.factorial(k)


def poisson_cdf(k, lam):
    """
    Calculate Poisson cumulative distribution function (CDF).
    
    P(X â¤ k) = sum of P(X = i) for i from 0 to k
    
    Parameters:
    k: number of events (non-negative integer)
    lam: average rate (lambda parameter, positive)
    
    Returns:
    Probability of k or fewer events
    """
    if k < 0:
        return 0
    if lam <= 0:
        raise ValueError("Lambda must be positive")
    
    cumulative_prob = sum(poisson_pmf(i, lam) for i in range(k + 1))
    return cumulative_prob


# Example usage and visualization
if __name__ == "__main__":
    # Parameters
    lambda_param = 3  # Average rate
    max_k = 12  # Maximum k value to plot
    
    # Calculate probabilities
    k_values = range(0, max_k + 1)
    pmf_values = [poisson_pmf(k, lambda_param) for k in k_values]
    cdf_values = [poisson_cdf(k, lambda_param) for k in k_values]
    
    # Print some example values
    print(f"Poisson Distribution with Î» = {lambda_param}")
    print("\nProbability Mass Function (PMF):")
    for k in range(13):
        prob = poisson_pmf(k, lambda_param)
        print(f"P(X = {k}) = {prob:.6f}")
    
    print("\nCumulative Distribution Function (CDF):")
    for k in range(13):
        prob = poisson_cdf(k, lambda_param)
        print(f"P(X â¤ {k}) = {prob:.6f}")
    
    # Create visualizations
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Plot PMF
    ax1.bar(k_values, pmf_values, color='steelblue', alpha=0.7, edgecolor='black')
    ax1.set_xlabel("X - axis is Number of Events (Success) k", fontsize=12)
    ax1.set_ylabel('Probability', fontsize=12)
    ax1.set_title(f'Poisson PMF (Î» = {lambda_param})', fontsize=14, fontweight='bold')
    #plt.xticks(k_values)
    ax1.grid(axis='y', alpha=0.3)
    
    # Plot CDF
    ax2.step(k_values, cdf_values, where='post', color='darkgreen', linewidth=2)
    ax2.scatter(k_values, cdf_values, color='darkgreen', s=50, zorder=5)
    ax2.set_xlabel("X - axis is Number of Events (Success) k", fontsize=12)
    ax2.set_ylabel('Cumulative Probability', fontsize=12)
    ax2.set_title(f'Poisson CDF (Î» = {lambda_param})', fontsize=14, fontweight='bold')
    plt.xticks(k_values)
    ax2.grid(alpha=0.3)
    ax2.set_ylim([0, 1.05])
    
    plt.tight_layout()
    plt.show()
```

```python
# Mean of the Discrete Poisson Probability Distribution
# Mean(Î¼) = n * p

Mean = n * p
print(f"Mean is lambda Î» = {lambda_param:.2f}")      # On Average 3 Number of Events (Succes) K will Occur

# Variance of the Discrete Binomial Probability Distribution
# Variance = n * p * q
Variance = n * p * (1 - p)
print(f"Variance is lambda Î» = {lambda_param:.2f}")     # Variance means outcome's average distance from the Mean

# Standard Deviation
Std_Dev = math.sqrt(lambda_param)
print(f"Std_Dev is {Std_Dev:.2f}")
```

### Hypergeometric Discrete Probability Distribution
* Binomial Distribution without Replacement is Called Hypergeometric Distribution
* Only 2 Outcomes, Success & Failure
* Successively (One by one) Without Replacement (Simultaneously)
* M is number of Success States in the Population
* N is total population
* n is the number of draws or trials
* k is desired number of Success
* Mean of the Hypergeometric Distribution E(X) = nM / N
* Variance of Hypergeometric Distribution = (nM / N) ( 1 â M/N ) ( N - n / N â 1 )

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
from scipy.stats import hypergeom

def hypergeometric_pmf(k, N, M, n):
    """
    Calculate hypergeometric probability mass function.
    
    Parameters:
    k : number of desired successes in the sample
    N : Total population size
    M : number of success states in population
    n : number of draws (sample size)
    
    Returns:
    probability of exactly k successes
    """
    numerator = comb(M, k, exact=True) * comb(N - M, n - k, exact=True)
    denominator = comb(N, n, exact=True)
    return numerator / denominator

def hypergeometric_cdf(k, N, M, n):
    """
    Calculate hypergeometric cumulative distribution function.
    
    Parameters:
    k : number of desired successes in the sample
    N : Total population size
    M : number of success states in population
    n : number of draws (sample size)
    
    Returns:
    probability of k or fewer successes
    """
    return sum(hypergeometric_pmf(i, N, M, n) for i in range(k + 1))

# Example: Drawing cards from a deck
# N = 52 cards total, M = 13 hearts, n = 5 cards drawn
N, M, n = 52, 13, 5

# Calculate probabilities for all possible outcomes
k_values = range(max(0, n - (N - M)), min(n, M) + 1)
pmf_values = [hypergeometric_pmf(k, N, M, n) for k in k_values]
cdf_values = [hypergeometric_cdf(k, N, M, n) for k in k_values]

# Verify using scipy
pmf_scipy = hypergeom.pmf(list(k_values), N, M, n)
cdf_scipy = hypergeom.cdf(list(k_values), N, M, n)

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# PMF plot
ax1.bar(k_values, pmf_values, alpha=0.7, label='Manual calculation')
ax1.plot(k_values, pmf_scipy, 'ro', markersize=8, label='scipy verification')
ax1.set_xlabel('Number of Hearts Successes (k)', fontsize=11)
ax1.set_ylabel('Probability', fontsize=11)
ax1.set_title(f'Hypergeometric PMF\nN={N}, M={M}, n={n}', fontsize=12)
ax1.legend()
ax1.grid(alpha=0.3)

# CDF plot
ax2.step(k_values, cdf_values, where='post', linewidth=2, label='Manual calculation')
ax2.plot(k_values, cdf_scipy, 'ro', markersize=8, label='scipy verification')
ax2.set_xlabel('Number of Hearts Successes (k)', fontsize=11)
ax2.set_ylabel('Cumulative Probability', fontsize=11)
ax2.set_title(f'Hypergeometric CDF\nN={N}, M={M}, n={n}', fontsize=12)
ax2.legend()
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Print some example calculations
print("Hypergeometric Distribution Example")
print(f"Population size (N): {N}")
print(f"Success states in population (M): {M}")
print(f"Sample size (n): {n}")
print("\n" + "="*50)

for k in k_values:
    pmf = hypergeometric_pmf(k, N, M, n)
    cdf = hypergeometric_cdf(k, N, M, n)
    print(f"P(X = {k}) = {pmf:.6f}")
    print(f"P(X â¤ {k}) = {cdf:.6f}")
    print("-" * 50)

# Calculate mean and variance
mean = n * M / N
variance = n * (M / N) * ((N - M) / N) * ((N - n) / (N - 1))
print(f"\nMean (expected value): {mean:.4f}")
print(f"Variance: {variance:.4f}")
print(f"Standard deviation: {np.sqrt(variance):.4f}")
```
