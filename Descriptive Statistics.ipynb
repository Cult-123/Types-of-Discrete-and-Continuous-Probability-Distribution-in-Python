# Types of Probability Distribution
1. Discrete Random Variable Probability Distribution
* Discrete Random Variables can take on a countable number of Possible Values
* We use Summation of Probability Mass Function
* Probability for an Exact Value of a Random Variable is Possible.
* Equation is Called PMF ie Probability Mass Function & P(x) >= 0, Î£P(x) = 1

2. Continuous Random Variable Probability Distribution
* We use intigration of Likelyhood Probability Density Function
* A random variable is continuous if it can be described by a Likelyhood Probability Density Function
* Probability for an Exact Value of Continuous Random Variable is not Possible
* Equation is Called Likelyhood PDF ie Probability Density Function $ \int_{-\infty}^{\infty} f(x) \,dx = 1 $
* f(x) likelyhood output value at any point of Continuous Random Variable x value, always will be $ f(x) >= 0 $

## Discrete Random Variable Probability Types

### Uniform & Cumulative Discrete Probability Distribution
* Throw a Dice Once
* Uniform Probability Distribution of A Dice (PMF)
* Cumulative Probability Distribution of A Dice (CDF)
* Die Roll is equivalent to Successively (One by one) With Replacement Method
* Trials are Independent

```python
import numpy as np
from scipy.stats import randint
import matplotlib.pyplot as plt

a, b = 1, 6
x = np.arange(a, b + 1)
dist = randint(a, b + 1)

pmf = dist.pmf(x)
cdf = dist.cdf(x)

# PMF plot
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, pmf, 'bo', ms=10, label='PMF')
plt.vlines(x, 0, pmf, colors='b', lw=4, alpha=0.7)
plt.title('PMF of Fair Six-Sided Die')
plt.xlabel('X axis is Sides of the Ludo Dice')
plt.ylabel('Y axis is Uniform Probability')
plt.xticks(x)
plt.ylim(0, 0.2)
plt.grid(True, alpha=0.3)
plt.legend()

# CDF plot (step style)
plt.subplot(1, 2, 2)
# Plot steps: horizontal line up to the point, then jump
plt.hlines(cdf, xmin=np.roll(x, 1), xmax=x, color='blue', alpha=0.7)
plt.plot(x, cdf, 'bo', ms=8)
plt.vlines(x, ymin=np.roll(cdf, 1), ymax=cdf, color='blue', alpha=0.7)

# Add the starting point at 0
plt.hlines(0, 0, 1, color='blue', alpha=0.7)
plt.title('CDF of Fair Six-Sided Die')
plt.xlabel('X axis is Sides of the Ludo Dice')
plt.ylabel('Y axis is Cumulative Probability')
plt.xticks(x)
plt.ylim(0, 1.05)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import math

# Mean of the Discrete Uniform Probability Distribution
# E(x) = Î£ x p(x)

Mean = sum((x * 1/6 ) for x in x)
print(f"Mean is {Mean}")      # means if we throw a Ludo Dice 1 time, on average 3.5 will come

# Var(x) = E(x^2) - [E(x)]^2
Var = sum(((x ** 2) * 1/6 ) for x in x) - (Mean) ** 2
print(f"Variance is {Var:.2f}")

# Standard Deviation
Std_Dev = math.sqrt(Var)
print(f"Std_Dev is {Std_Dev:.2f}")
```

* Coin Toss is equivalent to Successively (One by one) With Replacement Method
* Tossing a Dice for Appearing 6 or any other particular side as success is equivalent to Successively (One by one) With Replacement Method

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# --- Parameters ---
n = 10  # Total number of trials (e.g., coin flips)
p = 0.5 # Probability of success (e.g., getting heads)

# --- Calculate PMF for all possible outcomes (k = 0 to n) ---
# k: The number of successful outcomes you're interested in
k_values = list(range(n + 1))
probabilities = [binom.pmf(k, n, p) for k in k_values]

# --- Display Results ---
print(f"Binomial Probabilities for n={n}, p={p}:")
for k, prob in zip(k_values, probabilities):
    print(f"P(k={k}): {prob:.4f}")

# --- Optional: Plot the distribution (requires matplotlib) ---
plt.bar(k_values, probabilities)
plt.xlabel("X - axis is Number of Times Head (Successes (k)) Appears")
plt.ylabel("Y - axis is Binomial Probability")
plt.title(f"Binomial Distribution PMF (n={n}, p={p})")
plt.show()
```

```python
# Cumulatime Probability Distribution of the Above
Cumulative_Prob_dist = [binom.cdf(k, n, p) for k in k_values]

# --- Display Results ---
print(f"Binomial Cumulative Probabilities for n={n}, p={p}:")
for k, Cum_prob in zip(k_values, Cumulative_Prob_dist):
    print(f"P(k={k}): {Cum_prob:.4f}")

# plotting the graph
plt.ylabel('Y - axis is Binomial Cumulative Probability')
plt.xlabel('X - axis is Upto no of times Head (Successes (k)) Appears')
plt.title(f"Binomial Distribution CDF (n={n}, p={p})")
plt.bar(k_values, Cumulative_Prob_dist)
plt.show()
```

```python
# Mean of the Discrete Binomial Probability Distribution
# Mean(Î¼) = n * p

Mean = n * p
print(f"Mean is {Mean}")      # means if we Toss a Coin 10 times, on average 5 times Head will come

# Variance of the Discrete Binomial Probability Distribution
# The variance of a discrete probability distribution measures the spread of outcomes, calculated as the 
# Variance = n * p * q
Variance = n * p * (1 - p)
print(f"Variance is {Variance:.2f}")

# Standard Deviation
Std_Dev = math.sqrt(Variance)
print(f"Std_Dev is {Std_Dev:.2f}")
```

### Poisson Discrete Probability Distribution
* Poison Distribution is a kind of Binomial Distribution (Successively (One by one) With Replacement)
* Trials are Independent
* Only 2 Outcomes are Possible. Success & Failure.
* Where No of Trials are greater than 20
* Distribution mean Î¼ = np
* Variance (ÏÂ²) = np
* e = 2.71828
* n = number of trials
* p = prob of success is very small
* q = prob of failure is very big
* X = required no of success

<img src="DataFile/poisson.png"/>
<img src="DataFile/Poisson Binom Relation.png"/>

```python
import math
import matplotlib.pyplot as plt

def poisson_pmf(k, lam):
    """
    Calculate Poisson probability mass function (PMF).
    
    P(X = k) = (Î»^k * e^(-Î»)) / k!
    
    Parameters:
    k: number of events (non-negative integer)
    lam: average rate (lambda parameter, positive)
    
    Returns:
    Probability of exactly k events
    """
    if k < 0 or not isinstance(k, int):
        return 0
    if lam <= 0:
        raise ValueError("Lambda must be positive")
    
    return (lam ** k) * math.exp(-lam) / math.factorial(k)


def poisson_cdf(k, lam):
    """
    Calculate Poisson cumulative distribution function (CDF).
    
    P(X â¤ k) = sum of P(X = i) for i from 0 to k
    
    Parameters:
    k: number of events (non-negative integer)
    lam: average rate (lambda parameter, positive)
    
    Returns:
    Probability of k or fewer events
    """
    if k < 0:
        return 0
    if lam <= 0:
        raise ValueError("Lambda must be positive")
    
    cumulative_prob = sum(poisson_pmf(i, lam) for i in range(k + 1))
    return cumulative_prob


# Example usage and visualization
if __name__ == "__main__":
    # Parameters
    lambda_param = 3  # Average rate
    max_k = 12  # Maximum k value to plot
    
    # Calculate probabilities
    k_values = range(0, max_k + 1)
    pmf_values = [poisson_pmf(k, lambda_param) for k in k_values]
    cdf_values = [poisson_cdf(k, lambda_param) for k in k_values]
    
    # Print some example values
    print(f"Poisson Distribution with Î» = {lambda_param}")
    print("\nProbability Mass Function (PMF):")
    for k in range(13):
        prob = poisson_pmf(k, lambda_param)
        print(f"P(X = {k}) = {prob:.6f}")
    
    print("\nCumulative Distribution Function (CDF):")
    for k in range(13):
        prob = poisson_cdf(k, lambda_param)
        print(f"P(X â¤ {k}) = {prob:.6f}")
    
    # Create visualizations
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Plot PMF
    ax1.bar(k_values, pmf_values, color='steelblue', alpha=0.7, edgecolor='black')
    ax1.set_xlabel("X - axis is Number of Events (Success) k", fontsize=12)
    ax1.set_ylabel('Y - axis is Poisson Probability', fontsize=12)
    ax1.set_title(f'Poisson PMF (Î» = {lambda_param})', fontsize=14, fontweight='bold')
    #plt.xticks(k_values)
    ax1.grid(axis='y', alpha=0.3)
    
    # Plot CDF
    ax2.step(k_values, cdf_values, where='post', color='darkgreen', linewidth=2)
    ax2.scatter(k_values, cdf_values, color='darkgreen', s=50, zorder=5)
    ax2.set_xlabel("X - axis is Number of Events (Success) k", fontsize=12)
    ax2.set_ylabel('Y - axis is Cumulative Poisson Probability', fontsize=12)
    ax2.set_title(f'Poisson CDF (Î» = {lambda_param})', fontsize=14, fontweight='bold')
    plt.xticks(k_values)
    ax2.grid(alpha=0.3)
    ax2.set_ylim([0, 1.05])
    
    plt.tight_layout()
    plt.show()
```

```python
# Mean of the Discrete Poisson Probability Distribution
# Mean(Î¼) = n * p

Mean = n * p
print(f"Mean is lambda Î» = {lambda_param:.2f}")      # On Average 3 Number of Events (Succes) K will Occur

# Variance of the Discrete Binomial Probability Distribution
# Variance = n * p * q
Variance = n * p * (1 - p)
print(f"Variance is lambda Î» = {lambda_param:.2f}")     # Variance means outcome's average distance from the Mean

# Standard Deviation
Std_Dev = math.sqrt(lambda_param)
print(f"Std_Dev is {Std_Dev:.2f}")
```

### Hypergeometric Discrete Probability Distribution
* Binomial Distribution without Replacement is Called Hypergeometric Distribution
* Only 2 Outcomes are Possible. Success & Failure
* Successively (One by one) Without Replacement (Simultaneously)
* M = number of Success States in the Population
* N = total population
* n = the number of draws or trials
* k is desired number of Success
* Mean of the Hypergeometric Distribution E(X) = nM / N
* Variance of Hypergeometric Distribution = (nM / N) ( 1 â M/N ) ( N - n / N â 1 )

<img src="DataFile/Hypergeometric.png"/>
<img src="DataFile/Hypergeometric2.png"/>
<img src="DataFile/Hypergeom.png" style="height:324px; width:575px"/>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
from scipy.stats import hypergeom

def hypergeometric_pmf(k, N, M, n):
    """
    Calculate hypergeometric probability mass function.
    
    Parameters:
    k : number of desired successes in the sample
    N : Total population size
    M : number of success states in population
    n : number of draws (sample size)
    
    Returns:
    probability of exactly k successes
    """
    numerator = comb(M, k, exact=True) * comb(N - M, n - k, exact=True)
    denominator = comb(N, n, exact=True)
    return numerator / denominator

def hypergeometric_cdf(k, N, M, n):
    """
    Calculate hypergeometric cumulative distribution function.
    
    Parameters:
    k : number of desired successes in the sample
    N : Total population size
    M : number of success states in population
    n : number of draws (sample size)
    
    Returns:
    probability of k or fewer successes
    """
    return sum(hypergeometric_pmf(i, N, M, n) for i in range(k + 1))

# Example: Drawing cards from a deck
# N = 52 cards total, M = 13 hearts, n = 5 cards drawn
N, M, n = 52, 13, 5

# Calculate probabilities for all possible outcomes
k_values = range(max(0, n - (N - M)), min(n, M) + 1)
pmf_values = [hypergeometric_pmf(k, N, M, n) for k in k_values]
cdf_values = [hypergeometric_cdf(k, N, M, n) for k in k_values]

# Verify using scipy
pmf_scipy = hypergeom.pmf(list(k_values), N, M, n)
cdf_scipy = hypergeom.cdf(list(k_values), N, M, n)

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# PMF plot
ax1.bar(k_values, pmf_values, alpha=0.7, label='Manual calculation')
ax1.plot(k_values, pmf_scipy, 'ro', markersize=8, label='scipy verification')
ax1.set_xlabel('Number of Hearts Successes (k)', fontsize=11)
ax1.set_ylabel('Y - axis is Hypergeometric Probability', fontsize=11)
ax1.set_title(f'Hypergeometric PMF\nN={N}, M={M}, n={n}', fontsize=12)
ax1.legend()
ax1.grid(alpha=0.3)

# CDF plot
ax2.step(k_values, cdf_values, where='post', linewidth=2, label='Manual calculation')
ax2.plot(k_values, cdf_scipy, 'ro', markersize=8, label='scipy verification')
ax2.set_xlabel('Number of Hearts Successes (k)', fontsize=11)
ax2.set_ylabel('Y - axis is Cumulative Hypergeometric Probability', fontsize=11)
ax2.set_title(f'Hypergeometric CDF\nN={N}, M={M}, n={n}', fontsize=12)
ax2.legend()
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Print some example calculations
print("Hypergeometric Distribution Example")
print(f"Population size (N): {N}")
print(f"Success states in population (M): {M}")
print(f"Sample size (n): {n}")
print("\n" + "="*50)

for k in k_values:
    pmf = hypergeometric_pmf(k, N, M, n)
    cdf = hypergeometric_cdf(k, N, M, n)
    print(f"P(X = {k}) = {pmf:.6f}")
    print(f"P(X â¤ {k}) = {cdf:.6f}")
    print("-" * 50)

# Calculate mean and variance
mean = n * M / N
variance = n * (M / N) * ((N - M) / N) * ((N - n) / (N - 1))
print(f"\nMean (expected value): {mean:.4f}")
print(f"Variance: {variance:.4f}")
print(f"Standard deviation: {np.sqrt(variance):.4f}")
```

### Geometric Discrete Probability Distribution (Probability of getting first success at "n" th trial) (Binomial Format)
* Only 2 Outcomes are Possible, Success & Failure
* Trials are independent
* Successively (One by one) With Replacement
* Probability of First Success happens at 1st Trial, 2nd Trial, 3rd Trial etc.
* n is the number of draws or trials
* if first success happens at "n" th Trial
* Then first "n - 1" Trails are must be failures.
* The probability of success on the X-th trial = $(1 - P)^{X - 1} \,P$
* P(X = 4) = $(1 - P)^{4 - 1} \,P$
* P(X =< 3) = P(X = 1) + P(X = 2) + P(X = 3)   # Cumulative Probability of Success upto 3rd Trial

* Mean = $1 / P$
* Var = $(1 - P) / P^{2}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

def geometric_pmf(k, p):
    """
    Calculate geometric probability mass function.
    
    Parameters:
    k : number of trials until first success (k >= 1)
    p : probability of success on each trial (0 < p <= 1)
    
    Returns:
    probability of first success on the k-th trial
    """
    return (1 - p)**(k - 1) * p

def geometric_cdf(k, p):
    """
    Calculate geometric cumulative distribution function.
    
    Parameters:
    k : number of trials until first success (k >= 1)
    p : probability of success on each trial (0 < p <= 1)
    
    Returns:
    probability of first success within k trials
    """
    return 1 - (1 - p)**k

# Example: Rolling a die until you get a 6
# p = 1/6 (probability of rolling a 6)
p = 1/6

# Calculate probabilities for trials 1 to 20
k_values = range(1, 21)
pmf_values = [geometric_pmf(k, p) for k in k_values]
cdf_values = [geometric_cdf(k, p) for k in k_values]

# Verify using scipy
pmf_scipy = geom.pmf(list(k_values), p)
cdf_scipy = geom.cdf(list(k_values), p)

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# PMF plot
ax1.bar(k_values, pmf_values, alpha=0.7, label='Manual calculation')
ax1.plot(k_values, pmf_scipy, 'ro', markersize=6, label='scipy verification')
ax1.set_xlabel('First Success at Trial Number (k)', fontsize=11)
ax1.set_ylabel('Y - axis is Geometric Probability', fontsize=11)
ax1.set_title(f'Geometric PMF\np = {p:.4f}', fontsize=12)
ax1.legend()
ax1.grid(alpha=0.3)

# CDF plot
ax2.step(k_values, cdf_values, where='post', linewidth=2, label='Manual calculation')
ax2.plot(k_values, cdf_scipy, 'ro', markersize=6, label='scipy verification')
ax2.set_xlabel('First Succes within Trials Number (k)', fontsize=11)
ax2.set_ylabel('Y - axis is Cumulative Geometric Probability', fontsize=11)
ax2.set_title(f'Geometric CDF\np = {p:.4f}', fontsize=12)
ax2.legend()
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Print example calculations
print("Geometric Distribution Example")
print(f"Probability of success (p): {p:.4f}")
print(f"Scenario: Rolling a die until getting a 6")
print("\n" + "="*60)

for k in range(1, 21):
    pmf = geometric_pmf(k, p)
    cdf = geometric_cdf(k, p)
    print(f"P(First success on trial {k:2d}) = {pmf:.6f}")
    print(f"P(First success within {k:2d} trials) = {cdf:.6f}")
    print("-" * 60)

# Calculate mean and variance
mean = 1 / p
variance = (1 - p) / p**2
std_dev = np.sqrt(variance)

print(f"\nMean (expected number of trials): {mean:.4f}")
print(f"Variance: {variance:.4f}")
print(f"Standard deviation: {std_dev:.4f}")

# Additional useful probabilities
print("\n" + "="*60)
print("Additional Insights:")
print(f"Probability of success within 5 trials: {geometric_cdf(5, p):.4f}")
print(f"Probability of success within 10 trials: {geometric_cdf(10, p):.4f}")
print(f"Probability of needing more than 10 trials: {1 - geometric_cdf(10, p):.4f}")

# Compare different probability values
print("\n" + "="*60)
print("Comparison of different success probabilities:")
print(f"{'p':<10} {'Mean trials':<15} {'P(success within 5 trials)'}")
print("-" * 60)
for prob in [0.1, 0.2, 0.3, 0.5, 0.8]:
    mean_trials = 1 / prob
    prob_within_5 = geometric_cdf(5, prob)
    print(f"{prob:<10.1f} {mean_trials:<15.2f} {prob_within_5:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
from scipy.stats import multinomial
from itertools import product

def multinomial_pmf(x, n, p):
    """
    Calculate multinomial probability mass function.
    
    Parameters:
    x : array-like, observed counts for each category [x1, x2, ..., xk]
    n : total number of trials (should equal sum(x))
    p : array-like, probabilities for each category [p1, p2, ..., pk]
    
    Returns:
    probability of observing the specific outcome x
    """
    x = np.array(x)
    p = np.array(p)
    
    if not np.isclose(sum(x), n):
        raise ValueError("Sum of x must equal n")
    if not np.isclose(sum(p), 1.0):
        raise ValueError("Sum of probabilities must equal 1")
    
    # Calculate multinomial coefficient: n! / (x1! * x2! * ... * xk!)
    coeff = factorial(n)
    for xi in x:
        coeff /= factorial(xi)
    
    # Calculate probability: coeff * p1^x1 * p2^x2 * ... * pk^xk
    prob = coeff * np.prod(p ** x)
    
    return prob

def generate_all_outcomes(n, k):
    """
    Generate all possible outcomes for n trials with k categories.
    
    Parameters:
    n : number of trials
    k : number of categories
    
    Returns:
    list of all possible outcome vectors
    """
    outcomes = []
    
    def backtrack(remaining, current, pos):
        if pos == k - 1:
            current.append(remaining)
            outcomes.append(current[:])
            current.pop()
            return
        
        for i in range(remaining + 1):
            current.append(i)
            backtrack(remaining - i, current, pos + 1)
            current.pop()
    
    backtrack(n, [], 0)
    return outcomes

def multinomial_marginal_cdf(x_target, category_idx, n, p):
    """
    Calculate marginal CDF for a specific category.
    P(X_i <= x_target) where X_i is the count for category i.
    
    Parameters:
    x_target : threshold value for the target category
    category_idx : index of the category of interest
    n : total number of trials
    p : array-like, probabilities for each category
    
    Returns:
    cumulative probability
    """
    k = len(p)
    all_outcomes = generate_all_outcomes(n, k)
    
    cumulative_prob = 0.0
    for outcome in all_outcomes:
        if outcome[category_idx] <= x_target:
            cumulative_prob += multinomial_pmf(outcome, n, p)
    
    return cumulative_prob

# Example: Rolling a die 10 times
# Categories: 1, 2, 3, 4, 5, 6 (each with probability 1/6)
n = 10  # number of trials
k = 6   # number of categories
p = np.array([1/6] * 6)  # equal probabilities

# Example outcomes
print("Multinomial Distribution Example")
print(f"Number of trials (n): {n}")
print(f"Number of categories (k): {k}")
print(f"Probabilities: {p}")
print(f"Scenario: Rolling a fair die {n} times")
print("\n" + "="*70)

# Calculate probabilities for specific outcomes
example_outcomes = [
    [2, 2, 2, 2, 1, 1],  # relatively balanced
    [5, 1, 1, 1, 1, 1],  # one category dominates
    [3, 3, 2, 1, 1, 0],  # some categories not observed
]

print("\nPMF Examples (specific outcomes):")
print("-"*70)
for outcome in example_outcomes:
    prob = multinomial_pmf(outcome, n, p)
    print(f"Outcome {outcome}: P = {prob:.8f}")

# Verify with scipy
print("\nVerification with scipy:")
print("-"*70)
for outcome in example_outcomes:
    prob_manual = multinomial_pmf(outcome, n, p)
    prob_scipy = multinomial.pmf(outcome, n, p)
    print(f"Outcome {outcome}:")
    print(f"  Manual: {prob_manual:.8f}, Scipy: {prob_scipy:.8f}")

# Marginal distribution for first category (rolling a 1)
print("\n" + "="*70)
print("Marginal (Binomial) CDF for Category 0 (rolling a 1):")
print("-"*70)
for x in range(n + 1):
    cdf = multinomial_marginal_cdf(x, 0, n, p)
    print(f"P(X_0 â¤ {x:2d}) = {cdf:.6f}")

# Visualization: Simulate multinomial distribution
np.random.seed(42)
n_simulations = 10000
simulated_data = multinomial.rvs(n, p, size=n_simulations)

# Plot distribution of outcomes for each category
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i in range(k):
    ax = axes[i]
    counts, bins, patches = ax.hist(simulated_data[:, i], bins=range(n+2), 
                                     alpha=0.7, edgecolor='black', density=True)
    
    # Overlay theoretical marginal distribution (binomial)
    x_range = range(n + 1)
    from scipy.stats import binom
    theoretical = binom.pmf(x_range, n, p[i])
    ax.plot(x_range, theoretical, 'ro-', linewidth=2, markersize=6, 
            label='Theoretical (Binomial)')
    
    ax.set_xlabel(f'Count for Ludo Die Side {i+1}', fontsize=10)
    ax.set_ylabel('Probability', fontsize=10)
    ax.set_title(f'Marginal (Binomial) Distribution - Ludo Die Side {i+1}', fontsize=11)
    ax.legend()
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Calculate mean and covariance
mean = n * p
Variance = n * p * (1 - p)
print("\n" + "="*70)
print("Distribution Parameters:")
print("-"*70)
print(f"Mean vector (expected counts): {mean}")
print()
print(f"Variance vector: {Variance}")

# Covariance matrix
cov_matrix = np.zeros((k, k))
for i in range(k):
    for j in range(k):
        if i == j:
            cov_matrix[i, j] = n * p[i] * (1 - p[i])
        else:
            cov_matrix[i, j] = -n * p[i] * p[j]

print(f"\nCovariance matrix:")
print(cov_matrix)

# Example with different probabilities (loaded die)
print("\n" + "="*70)
print("Example with biased die:")
p_biased = np.array([0.3, 0.25, 0.2, 0.15, 0.07, 0.03])
outcome_biased = [4, 3, 2, 1, 0, 0]
prob_biased = multinomial_pmf(outcome_biased, n, p_biased)
print(f"Probabilities: {p_biased}")
print(f"Outcome {outcome_biased}: P = {prob_biased:.8f}")
```

### Negative Binomial Discrete Probability Distribution (Probability of getting "r" th success at "x" th trial)
* Only 2 Outcomes are Possible, Success & Failure
* Trials are independent
* Successively (One by one) With Replacement
* "x" is the number of draws or trials (Variable)
* "r" th succes always will be =< "x" th trial (Fixed)
* Probability of Success = P
* Probability of Failure = (1 - P)
* x = r, (r + 1), (r + 2), (r + 3)... trials upto infinity is possible
* Probability of "r"th Success happens at "r"th trials, (r + 1) trials, (r + 2) trials, (r + 3) trials... upto infinity trials is possible
* Then first "(x - 1)" trials must produce "(r - 1)" Successes.
* Formula for number of failures before the "r"th success $P(Y = y) = \binom{r + y - 1}{y} p^r (1-p)^y$
* The probability of "r"th success on the X-th trial = $P(X = x) = \binom{x - 1}{r - 1} p^r (1-p)^{x-r}$
* Cumulative Distribution Function for "r"th success (CDF) = $P(X \leq x) = \sum_{i=0}^{x} \binom{i + r - 1}{i} p^r (1-p)^i$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
from scipy.stats import nbinom

def negative_binomial_pmf(k, r, p):
    """
    Calculate negative binomial probability mass function.
    
    Formula: P(X=k) = C(k+r-1, k) * p^r * (1-p)^k
    
    Parameters:
    k : number of failures before the r-th success (k >= 0)
    r : number of successes (must be positive integer)
    p : probability of success on each trial (0 < p <= 1)
    
    Returns:
    probability of exactly k failures before r-th success
    """
    # Binomial coefficient C(k+r-1, k)
    coeff = comb(k + r - 1, k, exact=True)
    
    # Calculate probability
    prob = coeff * (p ** r) * ((1 - p) ** k)
    
    return prob

def negative_binomial_cdf(k, r, p):
    """
    Calculate negative binomial cumulative distribution function.
    
    Parameters:
    k : number of failures before the r-th success (k >= 0)
    r : number of successes (must be positive integer)
    p : probability of success on each trial (0 < p <= 1)
    
    Returns:
    probability of k or fewer failures before r-th success
    """
    return sum(negative_binomial_pmf(i, r, p) for i in range(k + 1))

# Example: Flipping a coin until getting 5 heads
# p = 0.5 (probability of heads), r = 5 (waiting for 5th success)
r = 5  # number of successes
p = 0.5  # probability of success

# Calculate probabilities for 0 to 20 failures (Tails)
k_values = range(0, 21)
pmf_values = [negative_binomial_pmf(k, r, p) for k in k_values]
cdf_values = [negative_binomial_cdf(k, r, p) for k in k_values]

# Verify using scipy
pmf_scipy = nbinom.pmf(list(k_values), r, p)
cdf_scipy = nbinom.cdf(list(k_values), r, p)

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# PMF plot
ax1.bar(k_values, pmf_values, alpha=0.7, label='Manual calculation')
ax1.plot(k_values, pmf_scipy, 'ro', markersize=6, label='scipy verification')
ax1.set_xlabel('Number of Failures (k) Before the 5th head Comes', fontsize=11)
ax1.set_ylabel('Y - axis is Negative Binomial Probability', fontsize=11)
ax1.set_title(f'Negative Binomial PMF\nr={r}, p={p}', fontsize=12)
ax1.legend()
ax1.grid(alpha=0.3)

# CDF plot
ax2.step(k_values, cdf_values, where='post', linewidth=2, label='Manual calculation')
ax2.plot(k_values, cdf_scipy, 'ro', markersize=6, label='scipy verification')
ax2.set_xlabel('Number of Failures (k) Before the 5th head Comes', fontsize=11)
ax2.set_ylabel('Y - axis is Cumulative Negative Binomial Probability', fontsize=11)
ax2.set_title(f'Negative Binomial CDF\nr={r}, p={p}', fontsize=12)
ax2.legend()
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Print example calculations
print("Negative Binomial Distribution Example")
print(f"Number of successes needed (r): {r}")
print(f"Probability of success (p): {p}")
print(f"Scenario: Flipping a coin until getting {r} heads")
print("\n" + "="*70)

for k in range(0, 11):
    pmf = negative_binomial_pmf(k, r, p)
    cdf = negative_binomial_cdf(k, r, p)
    total_trials = k + r
    print(f"P(exactly {k:2d} failures before {r}th success) = {pmf:.6f}")
    print(f"  (i.e., {r}th success occurs on trial {total_trials})")
    print(f"P({k:2d} or fewer failures) = {cdf:.6f}")
    print("-" * 70)

# Calculate mean and variance
mean = r * (1 - p) / p
variance = r * (1 - p) / (p ** 2)
std_dev = np.sqrt(variance)

print(f"\nMean (expected number of failures): {mean:.4f}")
print(f"Variance: {variance:.4f}")
print(f"Standard deviation: {std_dev:.4f}")

# Alternative interpretation: total trials until r-th success
print("\n" + "="*70)
print("Alternative interpretation (total trials):")
print(f"Expected total trials until {r}th success: {r/p:.4f}")

# Compare different parameter values
print("\n" + "="*70)
print("Effect of different parameters:")
print(f"{'r':<5} {'p':<8} {'Mean failures':<18} {'Std Dev':<15}")
print("-" * 70)

for r_val, p_val in [(3, 0.3), (5, 0.5), (10, 0.7), (5, 0.2), (5, 0.8)]:
    mean_val = r_val * (1 - p_val) / p_val
    std_val = np.sqrt(r_val * (1 - p_val) / (p_val ** 2))
    print(f"{r_val:<5} {p_val:<8.1f} {mean_val:<18.4f} {std_val:<15.4f}")

# Visualize effect of varying r and p
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Effect of varying r (fixed p)
ax1 = axes[0]
p_fixed = 0.5
k_range = range(0, 25)
for r_val in [3, 5, 10, 15]:
    pmf_vals = [negative_binomial_pmf(k, r_val, p_fixed) for k in k_range]
    ax1.plot(k_range, pmf_vals, marker='o', markersize=4, label=f'r={r_val}')

ax1.set_xlabel('Number of Failures (k) Before the "r"th head Comes', fontsize=11)
ax1.set_ylabel('Y - axis is Negative Binomial Probability', fontsize=11)
ax1.set_title(f'Effect of r on PMF (p={p_fixed})', fontsize=12)
ax1.legend()
ax1.grid(alpha=0.3)

# Effect of varying p (fixed r)
ax2 = axes[1]
r_fixed = 5
for p_val in [0.3, 0.5, 0.7, 0.9]:
    pmf_vals = [negative_binomial_pmf(k, r_fixed, p_val) for k in k_range]
    ax2.plot(k_range, pmf_vals, marker='o', markersize=4, label=f'p={p_val}')

ax2.set_xlabel('Number of Failures (k) as per different P values', fontsize=11)
ax2.set_ylabel('Y - axis is Negative Binomial Probability', fontsize=11)
ax2.set_title(f'Effect of p on PMF (r={r_fixed})', fontsize=12)
ax2.legend()
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Practical example: Quality control
print("\n" + "="*70)
print("Practical Example: Quality Control")
print("-" * 70)
print("A machine produces items with 10% defect rate (p=0.9 for good items)")
print("How many defective items before finding 20 good ones?")
print()

r_qc = 20  # need 20 good items
p_qc = 0.9  # probability of good item

mean_defects = r_qc * (1 - p_qc) / p_qc
print(f"Expected defective items: {mean_defects:.2f}")
print(f"Probability of 0 defects: {negative_binomial_pmf(0, r_qc, p_qc):.6f}")
print(f"Probability of 1 defect: {negative_binomial_pmf(1, r_qc, p_qc):.6f}")
print(f"Probability of 2 defects: {negative_binomial_pmf(2, r_qc, p_qc):.6f}")
print(f"Probability of 5 or fewer defects: {negative_binomial_cdf(5, r_qc, p_qc):.6f}")
```

### Multivariate Hypergeometric Discrete Probability Distribution (Probability of Getting more than 2 Outcomes)(Successively (One by one) Without Replacement (Simultaneously))
* More than 2 outcomes
* Successively (One by one) Without Replacement (Simultaneously)
* Trials are not Independent
* N = Total Population Size
* n = number of draws or trials

<img src="DataFile/Multivar Hypergeom2.png"/>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
from scipy.stats import multivariate_hypergeom
from itertools import product

def multivariate_hypergeometric_pmf(x, m, n):
    """
    Calculate multivariate hypergeometric probability mass function.
    
    Formula: P(X=x) = [C(m1,x1) * C(m2,x2) * ... * C(mk,xk)] / C(N,n)
    
    Parameters:
    x : array-like, number drawn from each category [x1, x2, ..., xk]
    m : array-like, number of items in each category [m1, m2, ..., mk]
    n : total number of items drawn (should equal sum(x))
    
    Returns:
    probability of drawing exactly x items from each category
    """
    x = np.array(x)
    m = np.array(m)
    
    if not np.isclose(sum(x), n):
        raise ValueError("Sum of x must equal n")
    
    if any(x > m):
        return 0.0  # Can't draw more items than available
    
    N = sum(m)  # Total population size
    
    # Calculate numerator: product of C(mi, xi) for all categories
    numerator = 1.0
    for mi, xi in zip(m, x):
        numerator *= comb(mi, xi, exact=True)
    
    # Calculate denominator: C(N, n)
    denominator = comb(N, n, exact=True)
    
    return numerator / denominator

def generate_all_multivariate_outcomes(n, m):
    """
    Generate all possible outcomes for multivariate hypergeometric.
    
    Parameters:
    n : number of items drawn
    m : array of items available in each category
    
    Returns:
    list of all valid outcome vectors
    """
    k = len(m)
    outcomes = []
    
    def backtrack(remaining, current, pos):
        if pos == k - 1:
            # Last category gets remaining items (if valid)
            if remaining <= m[pos]:
                current.append(remaining)
                outcomes.append(current[:])
                current.pop()
            return
        
        # Try all possible values for current category
        for i in range(min(remaining, m[pos]) + 1):
            current.append(i)
            backtrack(remaining - i, current, pos + 1)
            current.pop()
    
    backtrack(n, [], 0)
    return outcomes

def multivariate_hypergeometric_marginal_cdf(x_target, category_idx, m, n):
    """
    Calculate marginal CDF for a specific category.
    P(X_i <= x_target) where X_i is the count for category i.
    
    Parameters:
    x_target : threshold value for the target category
    category_idx : index of the category of interest
    m : array of items available in each category
    n : number of items drawn
    
    Returns:
    cumulative probability
    """
    all_outcomes = generate_all_multivariate_outcomes(n, m)
    
    cumulative_prob = 0.0
    for outcome in all_outcomes:
        if outcome[category_idx] <= x_target:
            cumulative_prob += multivariate_hypergeometric_pmf(outcome, m, n)
    
    return cumulative_prob

# Example: Drawing cards from a deck by suit
# m = [13, 13, 13, 13] (13 cards of each suit: hearts, diamonds, clubs, spades)
# n = 5 (drawing 5 cards)
m = np.array([13, 13, 13, 13])  # Items in each category
n = 5  # Number of items drawn
N = sum(m)  # Total population

print("Multivariate Hypergeometric Distribution Example")
print(f"Population composition: {m} (e.g., 13 hearts, 13 diamonds, 13 clubs, 13 spades)")
print(f"Total population size (N): {N}")
print(f"Number of items drawn (n): {n}")
print(f"Scenario: Drawing {n} cards from a standard deck")
print("\n" + "="*80)

# Example specific outcomes
example_outcomes = [
    [2, 2, 1, 0],  # 2 hearts, 2 diamonds, 1 club, 0 spades
    [5, 0, 0, 0],  # All 5 from one suit
    [1, 1, 1, 2],  # Mixed distribution
    [2, 1, 1, 1],  # Another mixed distribution
]

print("\nPMF Examples (specific outcomes):")
print("-"*80)
for outcome in example_outcomes:
    prob = multivariate_hypergeometric_pmf(outcome, m, n)
    print(f"Outcome {outcome}: P = {prob:.8f}")

# Verify with scipy
print("\nVerification with scipy:")
print("-"*80)
mv_hypergeom = multivariate_hypergeom(m, n)
for outcome in example_outcomes:
    prob_manual = multivariate_hypergeometric_pmf(outcome, m, n)
    prob_scipy = mv_hypergeom.pmf(outcome)
    print(f"Outcome {outcome}:")
    print(f"  Manual: {prob_manual:.8f}, Scipy: {prob_scipy:.8f}")

# Calculate marginal CDF for first category (hearts)
print("\n" + "="*80)
print("Marginal CDF for Category 0 (Hearts):")
print("-"*80)
for x in range(n + 1):
    cdf = multivariate_hypergeometric_marginal_cdf(x, 0, m, n)
    print(f"P(X_0 â¤ {x}) = {cdf:.6f}")

# Calculate mean and covariance
mean = n * m / N
print("\n" + "="*80)
print("Distribution Parameters:")
print("-"*80)
print(f"Mean vector (expected counts): {mean}")

# Covariance matrix
k = len(m)
cov_matrix = np.zeros((k, k))
for i in range(k):
    for j in range(k):
        if i == j:
            cov_matrix[i, j] = n * (m[i] / N) * (1 - m[i] / N) * (N - n) / (N - 1)
        else:
            cov_matrix[i, j] = -n * (m[i] / N) * (m[j] / N) * (N - n) / (N - 1)

print(f"\nCovariance matrix:")
print(cov_matrix)

# Generate all possible outcomes and their probabilities
print("\n" + "="*80)
print("Top 10 Most Likely Outcomes:")
print("-"*80)

all_outcomes = generate_all_multivariate_outcomes(n, m)
all_probs = [(outcome, multivariate_hypergeometric_pmf(outcome, m, n)) 
             for outcome in all_outcomes]
all_probs.sort(key=lambda x: x[1], reverse=True)

for i, (outcome, prob) in enumerate(all_probs[:10], 1):
    print(f"{i:2d}. {outcome} : P = {prob:.8f}")

print(f"\nTotal number of possible outcomes: {len(all_outcomes)}")
print(f"Sum of all probabilities: {sum(p for _, p in all_probs):.10f}")

# Visualization: Distribution of outcomes for each category
np.random.seed(42)
n_simulations = 10000
simulated_data = mv_hypergeom.rvs(size=n_simulations)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

category_names = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
for i in range(k):
    ax = axes[i]
    counts, bins, patches = ax.hist(simulated_data[:, i], bins=range(n+2), 
                                     alpha=0.7, edgecolor='black', density=True)
    
    # Overlay theoretical marginal distribution
    x_range = range(n + 1)
    theoretical = [multivariate_hypergeometric_marginal_cdf(x, i, m, n) - 
                   (multivariate_hypergeometric_marginal_cdf(x-1, i, m, n) if x > 0 else 0)
                   for x in x_range]
    ax.plot(x_range, theoretical, 'ro-', linewidth=2, markersize=6, 
            label='Theoretical')
    
    ax.set_xlabel(f'Count for {category_names[i]}', fontsize=10)
    ax.set_ylabel('Probability', fontsize=10)
    ax.set_title(f'Marginal Distribution - {category_names[i]}', fontsize=11)
    ax.legend()
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Example with different population composition
print("\n" + "="*80)
print("Example with Unequal Categories:")
print("-"*80)
print("Quality control: Box with 20 good, 10 average, 5 poor items")
m_qc = np.array([20, 10, 5])
n_qc = 8
N_qc = sum(m_qc)

print(f"Population: {m_qc} (good, average, poor)")
print(f"Drawing {n_qc} items")
print()

# Some interesting outcomes
outcomes_qc = [
    [5, 2, 1],  # Mostly good
    [3, 3, 2],  # Mixed
    [2, 3, 3],  # More poor items
]

for outcome in outcomes_qc:
    prob = multivariate_hypergeometric_pmf(outcome, m_qc, n_qc)
    print(f"Drawing {outcome[0]} good, {outcome[1]} average, {outcome[2]} poor: P = {prob:.6f}")

# Mean for this scenario
mean_qc = n_qc * m_qc / N_qc
print(f"\nExpected counts: {mean_qc}")
print(f"  Good: {mean_qc[0]:.2f}, Average: {mean_qc[1]:.2f}, Poor: {mean_qc[2]:.2f}")

# Visualize this example
fig, ax = plt.subplots(figsize=(10, 6))
all_outcomes_qc = generate_all_multivariate_outcomes(n_qc, m_qc)
all_probs_qc = [(outcome, multivariate_hypergeometric_pmf(outcome, m_qc, n_qc)) 
                for outcome in all_outcomes_qc]
all_probs_qc.sort(key=lambda x: x[1], reverse=True)

# Plot top 15 outcomes
top_outcomes = all_probs_qc[:15]
labels = [f"{o}" for o, _ in top_outcomes]
probs = [p for _, p in top_outcomes]

ax.barh(range(len(labels)), probs, alpha=0.7, edgecolor='black')
ax.set_yticks(range(len(labels)))
ax.set_yticklabels(labels, fontsize=9)
ax.set_xlabel('Probability', fontsize=11)
ax.set_ylabel('Outcome [Good, Average, Poor]', fontsize=11)
ax.set_title(f'Top 15 Most Likely Outcomes\nDrawing {n_qc} items from [{m_qc[0]}, {m_qc[1]}, {m_qc[2]}]', 
             fontsize=12)
ax.grid(alpha=0.3, axis='x')

plt.tight_layout()
plt.show()
```

### Bivariate Joint, Marginal & Conditional Discrete Probability Distribution. Also Covariance & Correlation Coefficient of Joint PMF
* This is Based on the Concept of Multivariate Hypergeometric Discrete Probability Distribution
* Mean of each Individual variable Mean E[X] and Mean E[Y]
* Combined Expectation of Bivariate Joint PMF is E[X, Y]
* Combined Covariance of Bivariate Joint PMF is Cov(X,Y)
* Combined Correlation Coefficient of Bivariate Joint PMF is Corr(X,Y)
* Variance of each Individual variable Var(X), Var(Y)
* Standard Deviation of each Individual variable SD(X), SD(Y)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm

class BivariateDiscrete:
    """
    Class to handle bivariate discrete probability distributions.
    Calculates joint, marginal, and conditional distributions.
    """
    
    def __init__(self, joint_pmf, x_values, y_values):
        """
        Initialize bivariate discrete distribution.
        
        Parameters:
        joint_pmf : 2D array, joint probability mass function P(X=x, Y=y)
        x_values : array, possible values of X
        y_values : array, possible values of Y
        """
        self.joint_pmf = np.array(joint_pmf)
        self.x_values = np.array(x_values)
        self.y_values = np.array(y_values)
        
        # Validate that probabilities sum to 1
        total_prob = np.sum(self.joint_pmf)
        if not np.isclose(total_prob, 1.0):
            raise ValueError(f"Joint PMF must sum to 1. Current sum: {total_prob}")
        
        # Calculate marginal distributions
        self.marginal_x = self._calculate_marginal_x()
        self.marginal_y = self._calculate_marginal_y()
    
    def _calculate_marginal_x(self):
        """Calculate marginal PMF of X: P(X=x) = Î£_y P(X=x, Y=y)"""
        return np.sum(self.joint_pmf, axis=1)
    
    def _calculate_marginal_y(self):
        """Calculate marginal PMF of Y: P(Y=y) = Î£_x P(X=x, Y=y)"""
        return np.sum(self.joint_pmf, axis=0)
    
    def conditional_x_given_y(self, y_value):
        """
        Calculate conditional PMF: P(X=x | Y=y)
        Formula: P(X=x | Y=y) = P(X=x, Y=y) / P(Y=y)
        """
        y_idx = np.where(self.y_values == y_value)[0][0]
        marginal_y_val = self.marginal_y[y_idx]
        
        if marginal_y_val == 0:
            raise ValueError(f"P(Y={y_value}) = 0, conditional distribution undefined")
        
        return self.joint_pmf[:, y_idx] / marginal_y_val
    
    def conditional_y_given_x(self, x_value):
        """
        Calculate conditional PMF: P(Y=y | X=x)
        Formula: P(Y=y | X=x) = P(X=x, Y=y) / P(X=x)
        """
        x_idx = np.where(self.x_values == x_value)[0][0]
        marginal_x_val = self.marginal_x[x_idx]
        
        if marginal_x_val == 0:
            raise ValueError(f"P(X={x_value}) = 0, conditional distribution undefined")
        
        return self.joint_pmf[x_idx, :] / marginal_x_val
    
    def expected_value_x(self):
        """Calculate E[X]"""
        return np.sum(self.x_values * self.marginal_x)
    
    def expected_value_y(self):
        """Calculate E[Y]"""
        return np.sum(self.y_values * self.marginal_y)
    
    def expected_value_xy(self):
        """
        Calculate E[XY] - the expected value of the product XY
        E[XY] = Î£_x Î£_y xy * P(X=x, Y=y)
        """
        exy = 0
        for i, x in enumerate(self.x_values):
            for j, y in enumerate(self.y_values):
                exy += x * y * self.joint_pmf[i, j]
        return exy
    
    def variance_x(self):
        """Calculate Var(X)"""
        ex = self.expected_value_x()
        return np.sum((self.x_values - ex)**2 * self.marginal_x)
    
    def variance_y(self):
        """Calculate Var(Y)"""
        ey = self.expected_value_y()
        return np.sum((self.y_values - ey)**2 * self.marginal_y)
    
    def covariance(self):
        """
        Calculate Cov(X,Y) = E[XY] - E[X]E[Y]
        """
        ex = self.expected_value_x()
        ey = self.expected_value_y()
        exy = self.expected_value_xy()
        
        return exy - ex * ey
    
    def correlation(self):
        """
        Calculate correlation coefficient: Ï = Cov(X,Y) / (Ï_X * Ï_Y)
        """
        cov = self.covariance()
        std_x = np.sqrt(self.variance_x())
        std_y = np.sqrt(self.variance_y())
        
        if std_x == 0 or std_y == 0:
            return 0
        
        return cov / (std_x * std_y)
    
    def independence_check(self, tolerance=1e-6):
        """
        Check if X and Y are independent.
        Independence: P(X=x, Y=y) = P(X=x) * P(Y=y) for all x, y
        """
        for i, x in enumerate(self.x_values):
            for j, y in enumerate(self.y_values):
                joint = self.joint_pmf[i, j]
                product = self.marginal_x[i] * self.marginal_y[j]
                if not np.isclose(joint, product, atol=tolerance):
                    return False
        return True

# Example 1: Rolling two dice
print("="*80)
print("Example 1: Rolling Two Dice")
print("="*80)
print("X = Result of first die, Y = Result of second die\n")

# Create joint PMF for two fair dice
x_vals = np.array([1, 2, 3, 4, 5, 6])
y_vals = np.array([1, 2, 3, 4, 5, 6])
joint_pmf_dice = np.ones((6, 6)) / 36  # All outcomes equally likely

dist_dice = BivariateDiscrete(joint_pmf_dice, x_vals, y_vals)

print("Joint PMF P(X=x, Y=y):")
header = "Y\\X"
print(f"{header:<6}", end='')
for x in x_vals:
    print(f"{x:<10}", end='')
print()
for j, y in enumerate(y_vals):
    print(f"{y:<6}", end='')
    for i, x in enumerate(x_vals):
        print(f"{joint_pmf_dice[i,j]:.6f}  ", end='')
    print()

print(f"\nMarginal PMF of X: {dist_dice.marginal_x}")
print(f"Marginal PMF of Y: {dist_dice.marginal_y}")
print(f"\n--- Expected Values ---")
print(f"E[X] = {dist_dice.expected_value_x():.4f}")
print(f"E[Y] = {dist_dice.expected_value_y():.4f}")
print(f"E[XY] = {dist_dice.expected_value_xy():.4f}")
print(f"\n--- Variances ---")
print(f"Var(X) = {dist_dice.variance_x():.4f}")
print(f"Var(Y) = {dist_dice.variance_y():.4f}")
print(f"SD(X) = {np.sqrt(dist_dice.variance_x()):.4f}")
print(f"SD(Y) = {np.sqrt(dist_dice.variance_y()):.4f}")
print(f"\n--- Covariance & Correlation ---")
print(f"Cov(X,Y) = {dist_dice.covariance():.4f}")
print(f"Correlation = {dist_dice.correlation():.4f}")
print(f"Independent? {dist_dice.independence_check()}")

# Conditional distribution
y_given = 3
cond_x_given_y = dist_dice.conditional_x_given_y(y_given)
print(f"\nConditional P(X=x | Y={y_given}):")
for i, x in enumerate(x_vals):
    print(f"  P(X={x} | Y={y_given}) = {cond_x_given_y[i]:.6f}")

# Example 2: Dependent variables (sum and difference)
print("\n" + "="*80)
print("Example 2: Dependent Variables")
print("="*80)
print("Two coins: X = number of heads, Y = number of tails\n")

# X can be 0, 1, or 2 (number of heads)
# Y can be 0, 1, or 2 (number of tails)
# But X + Y = 2 always (constraint)
x_vals_coins = np.array([0, 1, 2])
y_vals_coins = np.array([0, 1, 2])

# Joint PMF (many cells are zero due to constraint)
joint_pmf_coins = np.array([
    [0, 0, 0.25],   # X=0: must have Y=2 (TT)
    [0, 0.50, 0],   # X=1: must have Y=1 (HT or TH)
    [0.25, 0, 0]    # X=2: must have Y=0 (HH)
])

dist_coins = BivariateDiscrete(joint_pmf_coins, x_vals_coins, y_vals_coins)

print("Joint PMF P(X=x, Y=y):")
header = "Y\\X"
print(f"{header:<6}", end='')
for x in x_vals_coins:
    print(f"{x:<10}", end='')
print()
for j, y in enumerate(y_vals_coins):
    print(f"{y:<6}", end='')
    for i, x in enumerate(x_vals_coins):
        print(f"{joint_pmf_coins[i,j]:.6f}  ", end='')
    print()

print(f"\nMarginal PMF of X: {dist_coins.marginal_x}")
print(f"Marginal PMF of Y: {dist_coins.marginal_y}")
print(f"\n--- Expected Values ---")
print(f"E[X] = {dist_coins.expected_value_x():.4f}")
print(f"E[Y] = {dist_coins.expected_value_y():.4f}")
print(f"E[XY] = {dist_coins.expected_value_xy():.4f}")
print(f"\n--- Variances ---")
print(f"Var(X) = {dist_coins.variance_x():.4f}")
print(f"Var(Y) = {dist_coins.variance_y():.4f}")
print(f"SD(X) = {np.sqrt(dist_coins.variance_x()):.4f}")
print(f"SD(Y) = {np.sqrt(dist_coins.variance_y()):.4f}")
print(f"\n--- Covariance & Correlation ---")
print(f"Cov(X,Y) = {dist_coins.covariance():.4f}")
print(f"Correlation = {dist_coins.correlation():.4f}")
print(f"Independent? {dist_coins.independence_check()}")

# Conditional distributions
print("\nConditional Distributions:")
for y_val in [0, 1, 2]:
    cond = dist_coins.conditional_x_given_y(y_val)
    print(f"P(X=x | Y={y_val}): {cond}")

# Example 3: Custom distribution
print("\n" + "="*80)
print("Example 3: Custom Bivariate Distribution")
print("="*80)
print("Customer purchases: X = number of items, Y = satisfaction level\n")

x_vals_custom = np.array([1, 2, 3, 4])
y_vals_custom = np.array([1, 2, 3])  # 1=low, 2=medium, 3=high satisfaction

# Custom joint PMF (more items â higher satisfaction tendency)
joint_pmf_custom = np.array([
    [0.10, 0.08, 0.02],  # X=1
    [0.08, 0.15, 0.07],  # X=2
    [0.05, 0.18, 0.12],  # X=3
    [0.02, 0.08, 0.05]   # X=4
])

dist_custom = BivariateDiscrete(joint_pmf_custom, x_vals_custom, y_vals_custom)

print("Joint PMF P(X=x, Y=y):")
header = "Y\\X"
print(f"{header:<6}", end='')
for x in x_vals_custom:
    print(f"{x:<10}", end='')
print(f"{'Marginal Y':<12}")
print("-" * 60)
for j, y in enumerate(y_vals_custom):
    print(f"{y:<6}", end='')
    for i, x in enumerate(x_vals_custom):
        print(f"{joint_pmf_custom[i,j]:.6f}  ", end='')
    print(f"{dist_custom.marginal_y[j]:.6f}")
print("-" * 60)
print("Marg X", end='')
for i in range(len(x_vals_custom)):
    print(f"{dist_custom.marginal_x[i]:.6f}  ", end='')
print()

print(f"\n--- Expected Values ---")
print(f"E[Items] = {dist_custom.expected_value_x():.4f}")
print(f"E[Satisfaction] = {dist_custom.expected_value_y():.4f}")
print(f"E[Items Ã Satisfaction] = {dist_custom.expected_value_xy():.4f}")
print(f"\n--- Variances ---")
print(f"Var(Items) = {dist_custom.variance_x():.4f}")
print(f"Var(Satisfaction) = {dist_custom.variance_y():.4f}")
print(f"SD(Items) = {np.sqrt(dist_custom.variance_x()):.4f}")
print(f"SD(Satisfaction) = {np.sqrt(dist_custom.variance_y()):.4f}")
print(f"\n--- Covariance & Correlation ---")
print(f"Cov(Items, Satisfaction) = {dist_custom.covariance():.4f}")
print(f"Correlation = {dist_custom.correlation():.4f}")
print(f"Independent? {dist_custom.independence_check()}")

print("\nConditional P(Satisfaction | Items):")
for x_val in x_vals_custom:
    cond = dist_custom.conditional_y_given_x(x_val)
    print(f"  X={x_val} items: {[f'{p:.4f}' for p in cond]}")

# Visualizations
fig = plt.figure(figsize=(16, 12))

# 1. Joint PMF for dice (heatmap)
ax1 = plt.subplot(3, 3, 1)
sns.heatmap(joint_pmf_dice.T, annot=True, fmt='.4f', cmap='YlOrRd', 
            xticklabels=x_vals, yticklabels=y_vals, cbar_kws={'label': 'Probability'})
ax1.set_xlabel('X (First Die)')
ax1.set_ylabel('Y (Second Die)')
ax1.set_title('Joint PMF: Two Dice')

# 2. Marginal distributions for dice
ax2 = plt.subplot(3, 3, 2)
ax2.bar(x_vals - 0.2, dist_dice.marginal_x, width=0.4, alpha=0.7, label='P(X=x)')
ax2.bar(y_vals + 0.2, dist_dice.marginal_y, width=0.4, alpha=0.7, label='P(Y=y)')
ax2.set_xlabel('Value')
ax2.set_ylabel('Probability')
ax2.set_title('Marginal Distributions: Two Dice')
ax2.legend()
ax2.grid(alpha=0.3)

# 3. Conditional distribution for dice
ax3 = plt.subplot(3, 3, 3)
y_cond = 3
cond_vals = dist_dice.conditional_x_given_y(y_cond)
ax3.bar(x_vals, cond_vals, alpha=0.7, edgecolor='black')
ax3.set_xlabel('X')
ax3.set_ylabel('Probability')
ax3.set_title(f'Conditional P(X=x | Y={y_cond})')
ax3.grid(alpha=0.3)

# 4. Joint PMF for coins (heatmap)
ax4 = plt.subplot(3, 3, 4)
sns.heatmap(joint_pmf_coins.T, annot=True, fmt='.2f', cmap='Blues', 
            xticklabels=x_vals_coins, yticklabels=y_vals_coins, 
            cbar_kws={'label': 'Probability'})
ax4.set_xlabel('X (Heads)')
ax4.set_ylabel('Y (Tails)')
ax4.set_title('Joint PMF: Two Coins (Dependent)')

# 5. Marginal distributions for coins
ax5 = plt.subplot(3, 3, 5)
ax5.bar(x_vals_coins - 0.2, dist_coins.marginal_x, width=0.4, alpha=0.7, label='P(X=x)')
ax5.bar(y_vals_coins + 0.2, dist_coins.marginal_y, width=0.4, alpha=0.7, label='P(Y=y)')
ax5.set_xlabel('Value')
ax5.set_ylabel('Probability')
ax5.set_title('Marginal Distributions: Two Coins')
ax5.legend()
ax5.grid(alpha=0.3)

# 6. Conditional distributions for coins
ax6 = plt.subplot(3, 3, 6)
for y_val in y_vals_coins:
    cond = dist_coins.conditional_x_given_y(y_val)
    ax6.plot(x_vals_coins, cond, marker='o', label=f'Y={y_val}', linewidth=2)
ax6.set_xlabel('X')
ax6.set_ylabel('Probability')
ax6.set_title('Conditional P(X=x | Y=y) for Coins')
ax6.legend()
ax6.grid(alpha=0.3)

# 7. Joint PMF for custom (heatmap)
ax7 = plt.subplot(3, 3, 7)
sns.heatmap(joint_pmf_custom.T, annot=True, fmt='.2f', cmap='Greens', 
            xticklabels=x_vals_custom, yticklabels=y_vals_custom, 
            cbar_kws={'label': 'Probability'})
ax7.set_xlabel('X (Items)')
ax7.set_ylabel('Y (Satisfaction)')
ax7.set_title('Joint PMF: Customer Purchases')

# 8. Marginal distributions for custom
ax8 = plt.subplot(3, 3, 8)
x_pos = np.arange(len(x_vals_custom))
y_pos = np.arange(len(y_vals_custom))
ax8.bar(x_pos - 0.2, dist_custom.marginal_x, width=0.35, alpha=0.7, label='P(Items=x)')
ax8_twin = ax8.twinx()
ax8_twin.bar(y_pos + 0.2, dist_custom.marginal_y, width=0.35, alpha=0.7, 
             color='orange', label='P(Satisfaction=y)')
ax8.set_xlabel('Value')
ax8.set_ylabel('P(Items)', color='blue')
ax8_twin.set_ylabel('P(Satisfaction)', color='orange')
ax8.set_title('Marginal Distributions: Customer')
ax8.set_xticks(range(4))
ax8.set_xticklabels(['1', '2', '3', '1,2,3'])
ax8.grid(alpha=0.3)

# 9. Conditional for custom
ax9 = plt.subplot(3, 3, 9)
bar_width = 0.2
for i, x_val in enumerate(x_vals_custom):
    cond = dist_custom.conditional_y_given_x(x_val)
    positions = y_vals_custom + i * bar_width - bar_width * 1.5
    ax9.bar(positions, cond, width=bar_width, alpha=0.7, label=f'X={x_val}')
ax9.set_xlabel('Y (Satisfaction)')
ax9.set_ylabel('Probability')
ax9.set_title('Conditional P(Y=y | X=x)')
ax9.set_xticks(y_vals_custom)
ax9.legend()
ax9.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*80)
print("Summary of Key Concepts:")
print("="*80)
print("1. Joint PMF: P(X=x, Y=y) - probability of both events occurring")
print("2. Marginal PMF: P(X=x) = Î£_y P(X=x, Y=y)")
print("3. Conditional PMF: P(X=x | Y=y) = P(X=x, Y=y) / P(Y=y)")
print("4. Expected Values: E[X], E[Y], E[XY]")
print("5. Variances: Var(X) = E[XÂ²] - (E[X])Â², Var(Y) = E[YÂ²] - (E[Y])Â²")
print("6. Standard Deviations: SD(X) = âVar(X), SD(Y) = âVar(Y)")
print("7. Independence: P(X=x, Y=y) = P(X=x) * P(Y=y) for all x, y")
print("8. Covariance: Cov(X,Y) = E[XY] - E[X]E[Y]")
print("9. Correlation: Ï = Cov(X,Y) / (Ï_X * Ï_Y), ranges from -1 to 1")
```
